### Условия
Задача 6. Поиск аналитических решений для систем дифференциальных уравнений, описывающих конфликтные ситуации

⦁  Найти уравнение разделяющей прямой для фазового портрета системы, описывающей боевые потери регулярных частей.

⦁  Найти уравнение разделяющей прямой для фазового портрета системы, описывающей боевые потери при столкновении партизан.

⦁  Найти решение системы дифференциальных уравнений, описывающей боевые потери при столкновении партизан, в аналитическом виде.

⦁  Найти уравнение разделяющей кривой для фазового портрета системы, описывающей боевые потери при столкновениях регулярных частей с партизанами.

⦁  Найти решение системы дифференциальных уравнений, описывающей боевые потери при столкновениях регулярных частей с партизанами, в аналитическом виде.

### Шаг 1: Найти уравнение разделяющей прямой для фазового портрета системы, описывающей боевые потери регулярных частей.

Система дифференциальных уравнений:
$$
\begin{cases}
\frac{dx(t)}{dt} = -k_1 x(t)y(t) \\
\frac{dy(t)}{dt} = -k_2 x(t)
\end{cases}
$$

Для построения фазового портрета нужно выразить одно уравнение через другое. Для этого найдем $\frac{dy}{dx}$:

$$
\frac{dy}{dt} = -k_2 x \implies dt = -\frac{dy}{k_2 x}
$$

Подставим $dt$ в первое уравнение:

$$
\frac{dx}{dt} = -k_1 x y \implies \frac{dx}{dy} = \frac{dt}{dy} \cdot \frac{dx}{dt} = \left( -\frac{1}{k_2 x} \right) \cdot \left( -k_1 x y \right) = \frac{k_1 y}{k_2}
$$

Теперь найдем интеграл:

$$
\int \frac{dx}{dy} dy = \int \frac{k_1 y}{k_2} dy
$$
$$
x = \frac{k_1}{k_2} \cdot \frac{y^2}{2} + C
$$

Так как $C$ - интеграционная константа, изменяя ее, мы получим семейство кривых. Это уравнение можно считать уравнением разделяющей прямой для фазового портрета.

### Шаг 2: Найти уравнение разделяющей прямой для фазового портрета системы, описывающей боевые потери при столкновении партизан.

Для этого второго системы уравнений, если система идентична первой:

$$
\begin{cases}
\frac{dx}{dt} = -k_3 x y \\
\frac{dy}{dt} = -k_4 x
\end{cases}
$$

В аналогичной манере, разделяющая прямая может быть найдена так же:

$$
\frac{dy}{dt} = -k_4 x \implies dt = -\frac{dy}{k_4 x}
$$
$$
\frac{dx}{dt} = -k_3 x y \implies \frac{dx}{dy} = \frac{k_3 y}{k_4}
$$

Интегрирование: $x = \frac{k_3}{k_4} \cdot \frac{y^2}{2} + C$

### Шаг 3: Найти решение системы дифференциальных уравнений для столкновения партизан в аналитическом виде.

Рассмотрим:
$$
\begin{cases}
\frac{dx(t)}{dt} = -k_3 x(t)y(t) \\
\frac{dy(t)}{dt} = -k_4 x(t)
\end{cases}
$$

Рассмотрим систему:

$$
\begin{cases}
\frac{dx}{dt} = -k_3 x y \\
\frac{dy}{dt} = -k_4 x
\end{cases}
$$

$$
\frac{dy}{dx} = \frac{\frac{dy}{dt}}{\frac{dx}{dt}} = \frac{-k_4 x}{-k_3 x y} = \frac{k_4}{k_3 y}
$$

Интегрируем:

$$
\int y \, dy = \int \frac{k_4}{k_3} \, dx
$$
$$
\frac{y^2}{2} = \frac{k_4}{k_3} x + C
$$

Т.е. траектории будут: $y^2 = \frac{2 k_4}{k_3} x + 2C$

### Шаг 4: Найти уравнение разделяющей кривой для фазового портрета системы, описывающей боевые потери при столкновениях регулярных частей с партизанами.

Используем:

$$
\begin{cases}
\frac{dx}{dt} = -k_1 x y \\
\frac{dy}{dt} = -k_2 x
\end{cases}
$$

Находим $\frac{dy}{dx}$:

$$
\begin{cases}
\frac{dy}{dt} = -k_2 x \implies dt = -\frac{dy}{k_2 x} \\
\frac{dx}{dy} = \frac{-k_1 x y}{k_2 dt} = \frac{k_1 y}{k_2 x}
\end{cases}
$$

Интеграция:

$$\int x dx = \int \frac{k_2}{k_1} dy$$
$$\frac{x^2}{2} = \frac{k_2 y^2}{k_1} + C$$
$$x^2 = \frac{2 k_2}{k_1} y^2 + C$$

### Шаг 5: Найти решение системы дифференциальных уравнений для столкновений регулярных частей с партизанами в аналитическом виде.

Используем:

$$
\begin{cases}
\frac{dx(t)}{dt} = -k_1 x(t)y(t) \\
\frac{dy(t)}{dt} = -k_2 x(t)
\end{cases}
$$

Найдем решение системы с использованием метода переменных разделения.

Из второго уравнения: $\frac{dy}{dt} = -k_2 x$

Рассмотрим первый интеграл:

$y(t) = y_0 e^{-k_2 t}$
Подставляя $y(t)$ в первое уравнение: $dx = -k_1 x y_0 e^{-k_2 t} dt$

Решаем для $x(t)$: $\ln x = ln (x_0) – \frac{k_1 y_0}{k_2} e^{-k_2 t}$

Таким образом, $x(t)\text{ около }t \rightarrow \infty$.

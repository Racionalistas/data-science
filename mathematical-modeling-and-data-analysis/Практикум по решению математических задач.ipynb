### Задание 1

$$
\sqrt{x - 1} + x - 3 = 0
$$
### Решение
$$
\sqrt{x - 1} = 3 - x
$$
$$
(x - 1) = (3 - x)^2
$$
$$
x - 1 = 9 - 6x + x^2
$$
$$
x^2 - 7x + 10 = 0 
$$
$$
x = \frac{7 \pm \sqrt{49 - 40}}{2} = \frac{7 \pm 3}{2}
$$
$$
x_1 = 5 \quad \text{(не подходит, так как $\sqrt{5 - 1} + 5 - 3 \neq 0$)}
$$
$$
x_2 = 2 \quad \text{(подходит, так как $\sqrt{2 - 1} + 2 - 3 = 0$)}
$$

### Ответ
Единственный правильный ответ: $x = 2$

```python
# Проверка решения первой задачи
def solve_first_problem():
    x = symbols('x')
    equation = Eq(sqrt(x - 1) + x - 3, 0)
    solutions = solve(equation, x)
    solutions = [sol.evalf() for sol in solutions if sol.is_real]
    return solutions

first_problem_solution = solve_first_problem()
print(f"Решение первой задачи: {first_problem_solution}")
```

### Задание 2

$$
|2x + 1| + |x - 3| = 5
$$

### Решение

1. Случай 1: $ x \geq 3 $
$$
2x + 1 + x - 3 = 5
$$
$$
3x - 2 = 5

$$
3x = 7
$$
$$
x = \frac{7}{3} \quad (\text{не подходит, так как} \, \frac{7}{3} < 3)

2x + 1 + 3 - x = 5

x = 1 \quad (\text{подходит, так как} \, -0.5 \leq 1 < 3)

2. Случай 3: $ x < -0.5 $
$$
-(2x + 1) + 3 - x = 5

$$
-2x - 1 + 3 - x = 5
$$
$$
-3x + 2 = 5
$$
$$
-3x = 3

$$
x = -1 \quad (\text{подходит, так как} \, -1 < -0.5)

### Ответ
Два подходящих корня: $x = 1 \quad \text{и} \quad x = -1$

### Задание 3

$$
\begin{cases}
x^2 + y^2 = 25 \\
y^2 - x = 5
\end{cases}
$$

### Решение

1. Выразим $x$ из второго уравнения:
$$
y^2 - x = 5
$$
$$
x = y^2 - 5
$$

1. Подставим выражение для $x$ в первое уравнение:
$$
(y^2 - 5)^2 + y^2 = 25
$$
$$
y^4 - 10y^2 + 25 + y^2 = 25
$$
$$
y^4 - 9y^2 = 0
$$
$$
y^2(y^2 - 9) = 0
$$

2. Решим полученное уравнение, рассмотрим два случая:
$$
y^2 = 0 \quad \Rightarrow \quad y = 0
$$
$$
y^2 = 9 \quad \Rightarrow \quad y = \pm 3
$$

3. Найдем соответствующие значения для $x$:
$$
\text{Если} \; y = 0, \; x = 0^2 - 5 = -5
$$
$$
\text{Если} \; y = 3, \; x = 3^2 - 5 = 4
$$
$$
\text{Если} \; y = -3, \; x = (-3)^2 - 5 = 4
$$

### Ответ 
$(x, y) = (-5, 0) \quad \text{и} \quad (x, y) = (4, 3) \quad \text{и} \quad (x, y) = (4, -3)$

```python
# Проверка решения третьей задачи
def solve_third_problem():
    x, y = symbols('x y')
    eq1 = Eq(x**2 + y**2, 25)
    eq2 = Eq(y**2 - x, 5)
    solutions = solve((eq1, eq2), (x, y))
    solutions = [(sol[0].evalf(), sol[1].evalf()) for sol in solutions if sol[0].is_real and sol[1].is_real]
    return solutions

third_problem_solution = solve_third_problem()
print(f"Решение третьей задачи: {third_problem_solution}")

# График для третьей задачи
def plot_third_problem():
    x = np.linspace(-6, 6, 400)
    y = np.linspace(-6, 6, 400)
    X, Y = np.meshgrid(x, y)
    F = X**2 + Y**2 - 25
    G = Y**2 - X - 5
    
    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, F, [0], colors='r')
    plt.contour(X, Y, G, [0], colors='b')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Решение системы уравнений')
    plt.grid(True)
    plt.show()

plot_third_problem()
```

### Задание 4

Упростить выражение:
$$
\frac{1 + \sin\left(\frac{3\pi}{2} + 2x\right) - \sin^2 x}{\sin 2x \cdot \sin\left(\frac{\pi}{2} + x\right) + \sin x \cdot \cos(\pi - 2x)}
$$

### Решение
1. Используем тригонометрические тождества:
$
\sin\left(\frac{3\pi}{2} + 2x\right) = -\cos(2x)
$; 
$
\cos(\pi - 2x) = -\cos(2x)
$; 
$
\sin\left(\frac{\pi}{2} + x\right) = \cos(x)
$
$$
\frac{1 - \cos(2x) - \sin^2 x}{\sin 2x \cdot \cos(x) + \sin x \cdot (-\cos(2x))}
$$

2. Используем формулу приведения и тождества $\sin 2x = 2\sin x \cos x$ и $\cos(2x) = 1 - 2\sin^2 x$:

Числитель:
$$
1 - \cos(2x) - \sin^2 x = 1 - (1 - 2\sin^2 x) - \sin^2 x = 2\sin^2 x - \sin^2 x = \sin^2 x
$$

Знаменатель:
$$
\sin 2x \cdot \cos(x) + \sin x \cdot (-\cos(2x)) = 2\sin x \cos x \cdot \cos(x) - \sin x \cdot (-(1 - 2\sin^2 x)) = 2\sin x \cos^2 x + \sin x (1 - 2\sin^2 x)
$$
$$
2\sin x \cos^2 x - \sin x + 2\sin^3 x = \sin x (2\cos^2 x - 1 + 2\sin^2 x) = \sin x (2(\cos^2 x + \sin^2 x) - 1) = \sin x (2 \cdot 1 - 1) = \sin x
$$

### Ответ
$
\frac{\sin^2 x}{\sin x} = \sin x
$

```python
# Проверим решение четвертой задачи
def simplify_fourth_expression():
    x = symbols('x')
    numerator = 1 + sin(3 * pi / 2 + 2 * x) - sin(x) ** 2
    denominator = sin(2 * x) * sin(pi / 2 + x) + sin(x) * cos(pi - 2 * x)
    simplified_expression = simplify(numerator / denominator)
    return simplified_expression

fourth_problem_solution = simplify_fourth_expression()
print(f"Упрощенное выражение: {fourth_problem_solution}")
```

### Задание 5

Обозначим число через $ x $. Нам нужно максимизировать выражение $ x - x^2 $.

### Решение
1. Запишем функцию:
$$
f(x) = x - x^2
$$
$$
f'(x) = 1 - 2x
$$

3. Найдем критическую точку:
$$
f'(x) = 0
$$
$$
1 - 2x = 0
$$
$$
2x = 1
$$
$$
x = \frac{1}{2}
$$

4. Проверим, что это точка максимума (вторая производная):
$$
f''(x) = -2
$$

Так как $ f''(x) = -2 < 0 $, это точка максимума.

5. Найдем значение функции в этой точке:
$$
f\left(\frac{1}{2}\right) = \frac{1}{2} - \left(\frac{1}{2}\right)^2 = \frac{1}{2} - \frac{1}{4} = \frac{1}{4}
$$

### Ответ
Число $ x = \frac{1}{2} $ превышает свой квадрат на максимальное значение, равное $ \frac{1}{4} $.

```python
# Решение пятой задачи
def find_max_value():
    from scipy.optimize import minimize_scalar
    def f(x):
        return -(x - x**2)
    
    result = minimize_scalar(f, bounds=(0, 1), method='bounded')
    max_value = -result.fun
    best_x = result.x
    return best_x, max_value

fifth_problem_solution = find_max_value()
print(f"Наибольшее значение в пятой задаче достигается при x = {fifth_problem_solution[0]}, максимальное значение: {fifth_problem_solution[1]}")

def plot_fifth_problem():
    x = np.linspace(0, 1, 400)
    y = x - x**2
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label='y = x - x^2')
    plt.scatter(fifth_problem_solution[0], fifth_problem_solution[1], color='red', zorder=5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('График функции y = x - x^2')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_fifth_problem()
```

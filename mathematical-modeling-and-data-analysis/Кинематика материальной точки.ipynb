# Кинематика материальной точки

Кинематика изучает движение без учета причин, его вызывающих. Основными понятиями кинематики являются траектория, путь, перемещение, скорость и ускорение.

**Траектория** - это линия, которую описывает движущаяся материальная точка.\
**Путь** - это длина траектории, пройденная точкой.\
**Перемещение** - это вектор, соединяющий начальное и конечное положение точки.\
**Скорость** - это векторная величина, определяющая быстроту и направление движения точки. Формула средней скорости:
$
\vec{v}_{ср} = \frac{\vec{r}(t_2) - \vec{r}(t_1)}{t_2 - t_1}
$

Формула мгновенной скорости:
$
\vec{v}(t) = \frac{d\vec{r}(t)}{dt}
$

**Ускорение** - это векторная величина, характеризующая изменение скорости. Формула среднего ускорения:

$
\vec{a}_{ср} = \frac{\vec{v}(t_2) - \vec{v}(t_1)}{t_2 - t_1}
$

Формула мгновенного ускорения:
$
\vec{a}(t) = \frac{d\vec{v}(t)}{dt}
$

#### Пример 1: Равномерное прямолинейное движение

При равномерном движении скорость постоянна, ускорение равно нулю.

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры движения
v = 5  # скорость (м/с)
t = np.linspace(0, 10, 100)  # время от 0 до 10 секунд
r = v * t  # путь

# График зависимости пути от времени
plt.plot(t, r)
plt.xlabel('Время (с)')
plt.ylabel('Путь (м)')
plt.title('Равномерное прямолинейное движение')
plt.grid(True)
plt.show()
```

Пример 2: Равноускоренное движение
При равноускоренном движении ускорение постоянно.

```python
# Параметры движения
a = 2  # ускорение (м/с^2)
v0 = 0  # начальная скорость (м/с)
t = np.linspace(0, 10, 100)  # время от 0 до 10 секунд
r = v0 * t + 0.5 * a * t**2  # путь

# График зависимости пути от времени
plt.plot(t, r)
plt.xlabel('Время (с)')
plt.ylabel('Путь (м)')
plt.title('Равноускоренное движение')
plt.grid(True)
plt.show()
```

### Определение типа уравнения

Для определения типа данного уравнения:

$$
\frac{\partial^2 u}{\partial x^2} + 10 \frac{\partial^2 u}{\partial x \partial y} + 25 \frac{\partial^2 u}{\partial y^2} = 0
$$

мы рассмотрим обобщенное уравнение второго порядка в двух переменных:

$$
A \frac{\partial^2 u}{\partial x^2} + 2B \frac{\partial^2 u}{\partial x \partial y} + C \frac{\partial^2 u}{\partial y^2} = 0
$$

где \(A\), \(B\) и \(C\) являются коэффициентами при вторых производных.

Для данного уравнения коэффициенты равны:
- \(A = 1\)
- \(B = 5\) (поскольку \(2B = 10\))
- \(C = 25\)

Тип уравнения определяется дискриминантом:

$$
D = B^2 - AC
$$

Подставляя значения коэффициентов:

$$
D = 5^2 - 1 \cdot 25 = 25 - 25 = 0
$$

Следовательно:
- Если \(D > 0\), уравнение является гиперболическим.
- Если \(D = 0\), уравнение является параболическим.
- Если \(D < 0\), уравнение является эллиптическим.

В данном случае дискриминант равен нулю \(D = 0\), поэтому уравнение является **параболическим**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для отображения типа уравнения
def plot_pde_type(A, B, C, ax):
    D = B**2 - A*C
    if D > 0:
        equation_type = "Гиперболическое"
    elif D == 0:
        equation_type = "Параболическое"
    else:
        equation_type = "Эллиптическое"
    
    x = np.linspace(-10, 10, 400)
    y = np.linspace(-10, 10, 400)
    X, Y = np.meshgrid(x, y)
    Z = A*X**2 + 2*B*X*Y + C*Y**2
    
    # Выбираем уровни для контуров, сосредоточенные вокруг нуля
    levels = np.linspace(-3, 3, 200)
    contour = ax.contour(X, Y, Z, levels=levels, cmap='RdYlBu')
    ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

    levels = np.linspace(np.min(Z), np.max(Z), 50)
    contour = ax.contour(X, Y, Z, levels=levels, colors='r')
    ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

    ax.set_title(f"{equation_type}")
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)

# Создание рисунка
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Параболическое уравнение
plot_pde_type(1, 5, 25, axs[0])

# Гиперболическое уравнение
plot_pde_type(1, 2, -1, axs[1])

# Эллиптическое уравнение
plot_pde_type(1, 0, 1, axs[2])

plt.tight_layout()
plt.show()
```

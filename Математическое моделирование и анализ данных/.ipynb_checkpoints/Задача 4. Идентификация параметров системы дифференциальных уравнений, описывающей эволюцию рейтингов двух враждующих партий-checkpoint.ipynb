### Условия
Задача 4. Идентификация параметров системы дифференциальных уравнений, описывающей эволюцию рейтингов двух враждующих партий
Используя закодированную в электронной таблице Excel численную схему интегрирования системы дифференциальных уравнений (см. задачу 3), идентифицировать  коэффициенты k1 и k2, характеризующие эффективность средств пропаганды, по значениям наблюдаемых рейтингов (имеются в прилагаемом образце). Для численного решения задачи оптимизации использовать модуль «Поиск решения».

### Шаг 1: Определяем функции для вычисления производных и метод интеграции системы дифференциальных уравнений

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Заданные параметры
x0 = 10    # начальное значение x
y0 = 14    # начальное значение y
h = 0.01   # шаг интегрирования
t_end = 10 # конечное время интегрирования

# Количество шагов
n_steps = int(t_end / h)

# Функции для вычисления производных
def dx_dt(x, y, k1):
    return -k1 * y

def dy_dt(x, y, k2):
    return -k2 * x

# Метод Эйлера 2-го порядка (метод Хойна)
def integrate_system(k1, k2):
    t_values = np.zeros(n_steps + 1)
    x_values = np.zeros(n_steps + 1)
    y_values = np.zeros(n_steps + 1)

    x_values[0] = x0
    y_values[0] = y0

    for i in range(n_steps):
        t_values[i + 1] = t_values[i] + h

        x_predictor = x_values[i] + h * dx_dt(x_values[i], y_values[i], k1)
        y_predictor = y_values[i] + h * dy_dt(x_values[i], y_values[i], k2)

        x_values[i + 1] = x_values[i] + (h / 2) * (dx_dt(x_values[i], y_values[i], k1) + dx_dt(x_predictor, y_predictor, k1))
        y_values[i + 1] = y_values[i] + (h / 2) * (dy_dt(x_values[i], y_values[i], k2) + dy_dt(x_predictor, y_predictor, k2))

    return t_values, x_values, y_values
```

### Шаг 2: Задаем наблюдаемые данные 

```python
# Заданные наблюдаемые данные - нужно соответствовать размеру моделируемых данных
# Например, предположим что они сгенерированы для каждого временного шага
np.random.seed(0)  # для воспроизводимости
observed_x = 10 - 0.2 * np.arange(len(np.zeros(n_steps + 1))) + np.random.normal(scale=0.5, size=n_steps + 1)
observed_y = 14 - 0.3 * np.arange(len(np.zeros(n_steps + 1))) + np.random.normal(scale=0.5, size=n_steps + 1)
```

### Шаг 3: Вычисляем ошибку между моделируемыми и наблюдаемыми значениями

```python
def objective_function(params):
    k1, k2 = params
    _, modeled_x, modeled_y = integrate_system(k1, k2)
    error_x = np.sum((modeled_x - observed_x)**2)
    error_y = np.sum((modeled_y - observed_y)**2)
    return error_x + error_y

# Инициализация значений k1 и k2
initial_guess = [0.2, 0.8]
```

### Шаг 4: Находим оптимальные значения для коэффициентов

```python
# Использование метода минимизации
result = minimize(objective_function, initial_guess, method='Nelder-Mead')

# Оптимальные значения параметров
optimal_k1, optimal_k2 = result.x

# Печать результатов
print(f'Оптимальные значения k1: {optimal_k1}, k2: {optimal_k2}')
```

### Шаг 5: Визуализируем результаты моделирования по оптимальным значениям

```python
# Визуализация результатов
t_values, x_values, y_values = integrate_system(optimal_k1, optimal_k2)
plt.figure(figsize=(10, 5))
plt.plot(t_values, x_values, label='Моделируемый рейтинг партии x(t)')
plt.plot(t_values, y_values, label='Моделируемый рейтинг партии y(t)')
plt.plot(t_values, observed_x, 'o', label='Наблюдаемый рейтинг партии x(t)')
plt.plot(t_values, observed_y, 'o', label='Наблюдаемый рейтинг партии y(t)')
plt.xlabel('Время t')
plt.ylabel('Рейтинг')
plt.title('Сравнение моделируемых и наблюдаемых рейтингов двух враждующих партий')
plt.legend()
plt.grid()
plt.show()
```

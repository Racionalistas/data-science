Закодировать в электронной таблице Excel численную схему интегрирования системы дифференциальных уравнений, описывающей эволюцию рейтингов двух враждующих партий, методом Эйлера 2-го порядка. Интегрируемая система имеет вид: 

dx(t)/dt = - k1 y(t),
dy(t)/dt = - k2 x(t).
где x(t) и y(t) – рейтинги партий,  k1 и k2 - коэффициенты, характеризующие эффективность средств пропаганды. Рекомендуемые значения параметров: k1=0.26, k2=0.79, x(0)=10, y(0)=14, шаг интегрирования = 0.01.

### Задача 3. Численное интегрирование системы дифференциальных уравнений, описывающей эволюцию рейтингов двух враждующих партий

```python
import numpy as np
import matplotlib.pyplot as plt

# Заданные параметры
k1 = 0.26
k2 = 0.79
x0 = 10    # начальное значение x
y0 = 14    # начальное значение y
h = 0.01   # шаг интегрирования
t_end = 10 # конечное время интегрирования

# Количество шагов
n_steps = int(t_end / h)

# Массивы для хранения значений
t_values = np.zeros(n_steps + 1)
x_values = np.zeros(n_steps + 1)
y_values = np.zeros(n_steps + 1)

# Начальные условия
x_values[0] = x0
y_values[0] = y0

# Функции для вычисления производных
def dx_dt(x, y):
    return -k1 * y

def dy_dt(x, y):
    return -k2 * x
```

### Метод Эйлера 2-го порядка (метод Хойна) и визуализация результатов

```python
# Метод Эйлера 2-го порядка (метод Хойна)
for i in range(n_steps):
    t_values[i + 1] = t_values[i] + h

    x_predictor = x_values[i] + h * dx_dt(x_values[i], y_values[i])
    y_predictor = y_values[i] + h * dy_dt(x_values[i], y_values[i])

    x_values[i + 1] = x_values[i] + (h / 2) * (dx_dt(x_values[i], y_values[i]) + dx_dt(x_predictor, y_predictor))
    y_values[i + 1] = y_values[i] + (h / 2) * (dy_dt(x_values[i], y_values[i]) + dy_dt(x_predictor, y_predictor))

# Визуализация результатов
plt.figure(figsize=(10, 5))
plt.plot(t_values, x_values, label='Рейтинг партии x(t)')
plt.plot(t_values, y_values, label='Рейтинг партии y(t)')
plt.xlabel('Время t')
plt.ylabel('Рейтинг')
plt.title('Эволюция рейтингов двух враждующих партий')
plt.legend()
plt.grid()
plt.show()
```

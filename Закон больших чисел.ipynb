```python
import random
import matplotlib.pyplot as plt
```

```python
# функция для подбрасывания кости
def roll_dice():
    return random.randint(1, 6)
```

```python
# список для хранения результатов подбрасываний
results = []

# количество подбрасываний
num_rolls = 100_000
```

```python
# подбрасываем кость num_rolls раз и записываем результаты в список
for i in range(num_rolls):
    results.append(roll_dice())

# список для хранения сумм результатов подбрасываний до текущего момента
cumulative_sum = []

# переменная для хранения текущей суммы результатов подбрасываний
current_sum = 0

# вычисляем суммы результатов подбрасываний до каждого момента и записываем их в список
for result in results:
    current_sum += result
    cumulative_sum.append(current_sum / (1 + len(cumulative_sum)))

# вычисляем среднее значение результатов подбрасываний
mean = sum(results) / len(results)
```

```python
# выводим среднее значение и график, демонстрирующий закон больших чисел
print("Среднее значение: ", mean)
plt.plot(cumulative_sum)
plt.axhline(y=mean, color='r', linestyle='-')
plt.axhline(y=3.5, color='g', linestyle='-')
plt.show()
```

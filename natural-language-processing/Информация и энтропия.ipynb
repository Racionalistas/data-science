# Информация

**Информация** в теории информации измеряется как снижение неопределенности или неожиданность получения сообщения. Интуитивно, чем более неожиданным является сообщение, тем больше информации оно несет.

#### Мера информации

Информационным содержанием сообщения можно измерить с помощью **понятия вероятности**:

- Если событие $X$ с вероятностью $P(X)$ происходит, то информация, содержащаяся в этом событии, определяется как:

$$
I(X) = -\log_2(P(X))
$$

- Чем менее вероятно событие, тем больше информации оно несет. Например, информация от проведения честной подбрасывания монеты (вероятность 0.5) меньше, чем информация от менее вероятного события.

### Энтропия

**Энтропия** - это мера неопределенности или средней информации, присущей случайной переменной. Она является фундаментальной концепцией в теории информации и будет более понятна через несколько примеров.

#### Формула энтропии

Энтропия случайной переменной $X$ с дискретными исходами $x_1, x_2, \ldots, x_n$ и вероятностями $P(x_1), P(x_2), \ldots, P(x_n)$ определяется как:

$$
H(X) = -\sum_{i=1}^{n} P(x_i) \log_2(P(x_i))
$$

- Если все исходы событий равномерно распределены, энтропия будет максимальной.
- Если одно событие имеет вероятность 1 (то есть оно гарантировано произойдет), а остальные — 0, то энтропия будет минимальной (равной 0).

### Примеры

1. **Честная подбрасывание монеты**:

- Вероятности: $P(Heads) = 0.5$, $P(Tails) = 0.5$
- Энтропия: 

$$
H(X) = - (0.5 \log_2(0.5) + 0.5 \log_2(0.5)) = 1 \text{ бит}
$$

2. **Нечестное подбрасывание монеты**:

- Вероятности: $P(Heads) = 0.8$, $P(Tails) = 0.2$
- Энтропия:

$$
H(X) = - (0.8 \log_2(0.8) + 0.2 \log_2(0.2)) \approx 0.72 \text{ бита}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def entropy(probabilities):
    """
    Вычисление энтропии по заданным вероятностям событий.
    """
    probabilities = np.asarray(probabilities)
    # Маска для игнорирования нулевых вероятностей
    mask = probabilities > 0
    return -np.sum(probabilities[mask] * np.log2(probabilities[mask]), axis=-1)

# Генерация значений вероятности орла (P(heads))
prob_heads = np.linspace(0, 1, 1000)

# Вероятность орла и решки
prob_tails = 1 - prob_heads

# Объединение вероятностей орла и решки в одну матрицу
probabilities = np.vstack([prob_heads, prob_tails]).T

# Вычисление энтропии для каждой пары вероятностей
entropies = np.array([entropy(p) for p in probabilities])

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(prob_heads, entropies, color='blue', label='Энтропия')
plt.xlabel('Вероятность Орла (P(орел))')
plt.ylabel('Энтропия (биты)')
plt.title('Энтропия в зависимости от вероятности Орла')
plt.axvline(x=0.5, color='red', linestyle='--', label='Честная монета (P(орел) = 0.5)')
plt.legend()
plt.grid(True)
plt.show()
```

Этот код вычисляет энтропию для двух наборов вероятностей: честной и нечестной монеты.

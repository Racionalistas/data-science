**Задача 4: Нахождение особых точек системы**

$$
\begin{cases}
x' = y^2 - 4x^2 \newline
y' = 4y - 8
\end{cases} \quad \Longrightarrow \quad
\begin{cases}
y^2 - 4x^2 = 0 \newline
4y - 8 = 0
\end{cases}
$$
$$
4y = 8; \quad
y = 2 \quad \Longrightarrow \quad
y^2 - 4x^2 = 0; \quad
2^2 - 4x^2 = 0; \quad
4x^2 = 4; \\
x = \pm 1
$$

Особые точки: $(1; 2), (-1; 2)$

**1. Уравнения перехода к новым координатам для $(1; 2)$:**

$$
\begin{cases}
u = x - 1 \newline
v = y - 2
\end{cases} \quad \Longrightarrow \quad
\begin{cases}
x = u + 1 \newline
y = v + 2
\end{cases}
$$
$$
\frac{du}{dt}=(v+2)^2-4(u+1)^2=v^2+4v+4-4u^2-8u-4=-8u+4v
$$
$$
\frac{dv}{dt}=4(v+2)-8=4v+8-8=4v
$$

Первая особая точка:

$$
\begin{pmatrix}
-8 & 4 \newline
0 & 4 
\end{pmatrix}
\begin{bmatrix}
-8-\lambda & 4 \newline
0 & 4-\lambda 
\end{bmatrix}
=(-8-\lambda)(4-\lambda)-4*0=-32+8\lambda-4\lambda+\lambda^2
=\lambda^2+4\lambda-32 
$$
$$
D=4^2-4*1(-32)=16+128=144
$$
$$
\lambda_{1}=\frac{-4 + 12}{2}=4
$$
$$
\lambda_{2}=\frac{-4 - 12}{2}=-8
$$

Особая точка типа седло.\
Первый собственный вектор: 

a. 

$$
\begin{pmatrix}
-8 & 4 \newline
0 & 4 
\end{pmatrix}
\begin{pmatrix}
u \\ v 
\end{pmatrix}
=4
\begin{pmatrix}
u \\ v 
\end{pmatrix} 
$$
$$
u=1;\quad -8+4v=4;\quad 4v=12;\quad v=3; 
$$
$$
(1; 3)
$$

b.

$$
\begin{pmatrix}
-8 & 4 \newline
0 & 4 
\end{pmatrix}
\begin{pmatrix}
u \\ v 
\end{pmatrix}
=-8
\begin{pmatrix}
u \\ v 
\end{pmatrix} 
$$
$$
u=1;\quad -8+4v=-8;\quad 4v=0;\quad v=0; 
$$
$$
(1; 0)
$$

**2. Уравнения перехода к новым координатам для $(-1; 2)$:**

$$
\begin{cases}
u = x + 1 \newline
v = y - 2
\end{cases} \quad \Longrightarrow \quad
\begin{cases}
x = u - 1 \newline
y = v + 2
\end{cases} 
$$
$$
\frac{du}{dt} = (v+2)^2-4(u-1)^2=v^2+4v+4-4u^2+8u-4=8u+4v 
$$
$$
\frac{dv}{dt} = 4(v+2)-8=4v+8-8=4v
$$

Вторая особая точка:

$$
\begin{pmatrix}
8 & 4 \newline
0 & 4 
\end{pmatrix}
\begin{bmatrix}
8-\lambda & 4 \newline
0 & 4-\lambda 
\end{bmatrix}
=(8-\lambda)(4-\lambda)-4*0=32-8\lambda-4\lambda+\lambda^2
=\lambda^2+12\lambda+32
$$
$$
D=(-12)^2-4*1(-32)=144-128=16
$$
$$
\lambda_{1}=\frac{12+4}{2}=8
$$
$$
\lambda_{2}=\frac{12-4}{2}=4
$$

Особая точка типа узел.\
Второй собственный вектор: 

a. 

$$
\begin{pmatrix}
8 & 4 \newline
0 & 4 
\end{pmatrix}
\begin{pmatrix}
u \\ v 
\end{pmatrix}
=8
\begin{pmatrix}
u \\ v 
\end{pmatrix} 
$$
$$
u=1;\quad 8+4v=8;\quad 4v=0;\quad v=0;
$$
$$
(1; 0)
$$

$$

8 & 4 \newline
